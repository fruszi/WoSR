## 4. Model evaluation and tuning

```python
# import packages
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.model_selection import GridSearchCV

import time
```

### Cross-validation (Kfolds)

```python
train_Ycat = np.squeeze(train_Ycat)                                       # reshape labels to (rows,) from (rows,1)

t0 = time.time()

reg_cv_predictions = cross_val_predict(log_reg, train_X, train_Ycat, cv=5)
reg_cv_conf = confusion_matrix(train_Ycat, reg_cv_predictions)

svc_cv_predictions = cross_val_predict(svc, train_X, train_Ycat, cv=5)
svc_cv_conf = confusion_matrix(train_Ycat, svc_cv_predictions)

sgd_cv_predictions = cross_val_predict(sgd_clf, train_X, train_Ycat, cv=5)
sgd_cv_conf = confusion_matrix(train_Ycat, sgd_cv_predictions)

tree_cv_predictions = cross_val_predict(tree_clf, train_X, train_Ycat, cv=5)
tree_cv_conf = confusion_matrix(train_Ycat, tree_cv_predictions)

forest_cv_predictions = cross_val_predict(forest_clf, train_X, train_Ycat, cv=5)
forest_cv_conf = confusion_matrix(train_Ycat, forest_cv_predictions)

xgb_cv_predictions = cross_val_predict(xgb_clf, train_X, train_Ycat, cv=5)
xgb_cv_conf = confusion_matrix(train_Ycat, xgb_cv_predictions)

t1 = time.time()
print(t1 - t0)
```

##### Plot confusion matrices (with F1 scores) for comparison

```python
# function to construct confusion matrices, for plotting
def conf_matrix_plot(Y, conf_matrix, title, f1):
    classes = np.unique(Y)
    conf_norm = conf_matrix.astype('float') / conf_matrix.sum(axis=1)[:,np.newaxis]                     # normalized matrix
    f = plt.imshow(conf_norm, cmap=plt.cm.Reds, interpolation='None', vmin=0.0, vmax=1.0)
    plt.grid(False)
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes)
    plt.yticks(tick_marks, classes)
    fmt = '.2f' 
    thresh = conf_norm.max() / 2.
    for i, j in itertools.product(range(conf_norm.shape[0]), range(conf_norm.shape[1])):
        plt.text(j, i, format(conf_norm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if conf_norm[i, j] > thresh else "black")

    plt.title(title + ' :' + ' F1 = %.2f%%' % (f1*100) )
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    return(f)
```

```python
# name variables
matrx =  (reg_cv_conf,        svc_cv_conf,        sgd_cv_conf,        tree_cv_conf,        forest_cv_conf,      xgb_cv_conf)
preds =  (reg_cv_predictions,svc_cv_predictions,sgd_cv_predictions,tree_cv_predictions,forest_cv_predictions,xgb_cv_predictions)
titles = ('reg_cv',           'svc_cv',           'sgd_cv',           'tree_cv',           'forest_cv',          'xgb_cv')

# calculate F1 scores
f1 = list()
for mx in np.arange(len(matrx)):
    f1.append(precision_recall_fscore_support(train_Ycat, preds[mx], average='weighted')[2]) 

# plot confusion matrices, for all models 
for mx in np.arange(len(matrx)):
    plt.subplot(2, 3, mx+1)
    conf_matrix_plot(train_Y, matrx[mx], titles[mx], f1[mx])
```

##### Plot learning curves 
runs for >10 mins

```python
from sklearn.model_selection import learning_curve

def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None,
                        n_jobs=-1, train_sizes=np.linspace(.1, 1.0, 5)): 
                           
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    
    return(train_sizes, train_scores_mean, train_scores_std, test_scores_mean, test_scores_std, ylim, title)
```

```python
## plot learning curves, for all models 
estimators = (log_reg,   svc,  sgd_clf, tree_clf,   forest_clf, xgb_clf)
modelnames = ('log.reg','SVC', 'SGD',   'tree',     'forest',   'XGB')

t0 = time.time()
for mx in np.arange(len(estimators)):
    plt.subplot(3, 2, mx+1)
      
    # calculating learning curve parameters
    train_sizes, train_scores_mean, train_scores_std, test_scores_mean, test_scores_std, ylim, title = plot_learning_curve(
        estimators[mx], modelnames[mx], train_X, train_Ycat, ylim=(.7, 1.05), cv=None, 
        n_jobs=-1, train_sizes=np.linspace(.1, 1.0, 6))
    # ......................................
    
    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    #plt.xlabel("Training examples")
    plt.ylabel("Score")
    plt.grid(True)
    
    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="g")
    
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g", label="Cross-validation score")
    plt.legend(loc="best")
    
print(time.time() - t0)
save_fig("learning_curves")
```

### Tuning the selected model

```python
# construct an iterator
cv = StratifiedShuffleSplit(n_splits=10, test_size=0.2, random_state=2)
```

##### this step takes a long time... (hours)

best model is described in the cell below, no need to rerun gridsearch every time. 

```python
param_grid = [{'C': [1, 10, 100, 1000], 'gamma': [0.001, 0.0001], 'kernel': ['rbf'], 'class_weight': ['balanced']}]

t0 = time.time()
classifier = GridSearchCV(estimator=svc, cv=cv, param_grid=param_grid)
classifier.fit(train_X, train_Ycat)
print(time.time() - t0)
```

## Finally, classify!

#### Apply preprocessing steps to test set

```python
props, feature_count, cols = basic_data_properties(test_set)

test_set = test_set[cols[6:]]

# replace DIV errors with nan
test_set = test_set.replace('#DIV/0!',np.nan)

# convert all but class label
test_set.iloc[:,1:-1] = test_set.iloc[:,1:-1].convert_objects(convert_numeric=True)

# drop features with >95% missing values
features_missing_values = test_set.isnull().sum() / len(test_set) > .95
features_drop = list(features_missing_values[features_missing_values].index)
test_set.drop(features_drop, axis=1, inplace=True)
```

```python
test_Y = pd.DataFrame(test_set["classe"])
test_X = test_set.drop("classe", axis=1) 

# imputer
test_X = pd.DataFrame(imputer.fit_transform(test_X), columns=test_X.columns,
                       index = list(test_X.index.values))                           # returns array, not dataframe

# scaling
test_X = pd.DataFrame(scaler.fit_transform(test_X), columns=test_X.columns,
                       index = list(test_X.index.values))  

# label encoding
test_Ycat = pd.DataFrame(lb.fit_transform(test_Y), columns=test_Y.columns,
                         index = list(test_Y.index.values))
```

### Classify, and check outcome

```python
test_predictions = classifier.best_estimator_.predict(test_X)
conf = confusion_matrix(test_Ycat, test_predictions)

classes = np.unique(train_Y)
conf_norm = conf.astype('float') / conf.sum(axis=1)[:,np.newaxis]                                    # plot normalized matrix

plt.imshow(conf_norm, cmap=plt.cm.Reds, interpolation='None', figsize=(7,7))
#plt.colorbar()
plt.grid(False)
tick_marks = np.arange(len(classes))
plt.xticks(tick_marks, classes)
plt.yticks(tick_marks, classes)
fmt = '.2f' 
thresh = conf_norm.max() / 2.
for i, j in itertools.product(range(conf_norm.shape[0]), range(conf_norm.shape[1])):
    plt.text(j, i, format(conf_norm[i, j], fmt),
             horizontalalignment="center",
             color="white" if conf_norm[i, j] > thresh else "black")

#plt.tight_layout()
plt.ylabel('True label')
plt.xlabel('Predicted label')

plt.show()
```

```python
f1 = precision_recall_fscore_support(test_Ycat, test_predictions, average='weighted')

print("Precision (accuracy): %.2f%%" % (f1[0] * 100.0))
print("Recall (sensitivity): %.2f%%" % (f1[1] * 100.0))
print("F1 score: %.2f%%" % (f1[2] * 100.0))
```
